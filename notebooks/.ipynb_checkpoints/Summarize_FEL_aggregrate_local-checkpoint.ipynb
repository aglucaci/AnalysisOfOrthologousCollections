```python
import pandas as pd
import numpy as np
import os
import json
import altair as alt
import glob
from altair_saver import save
```

```python
FEL_RESULT_JSONS = glob.glob(os.path.join("..", "results", "mammalian_REV3L", "*.FEL.json"))

print(FEL_RESULT_JSONS)
```

```python
sorted(FEL_RESULT_JSONS)
```

```python
def getFELData(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["MLE"]["content"]["0"]
#end method

def getFELHeaders(json_file):
    with open(json_file, "r") as in_d:
        json_data = json.load(in_d)
    return json_data["MLE"]["headers"]
#end method
```

```python
order = range(1, len(FEL_RESULT_JSONS) + 1)
order
```

```python
df_holder = []
for i in order:
    #print(i)
    file = os.path.join("..", "results", "mammalian_REV3L", "mammalian_REV3L." + str(i) + ".codon.fas.FEL.json")
    print("# Opening;", file)
    columns = getFELHeaders(file)
    headers = [x[0] for x in columns]
    #headers
    #data = getFELData(JSON_FILE)
    df = pd.DataFrame(getFELData(file), columns=headers, dtype = float)
    #df["omega"] = df["beta"] / df["alpha"]
    #df.index += 1
    #df["Site"] = df.index
    #df
    df_holder.append(df)
    
```

```python
df_main = pd.concat(df_holder)
df_main = df_main.reset_index(drop=True)
df_main.index += 1
df_main["Site"] = df_main.index
#df_main.to_csv("AOC-REV3L_FELResults_all.csv")
df_main
```

```python
#df_main = df_main.rename(columns={"dN/dS MLE": "dNdS_MLE"})
df_main = df_main.rename(columns={"p-value": "p_value"})
#df_main
```

```python
df_main.columns
```

```python
#pd.to_numeric(df_main)
df_main = df_main.astype(float)
df_main
```

```python
source = df_main

line = alt.Chart(source).mark_circle().encode(
    x='Site',
    y = alt.Y('dN/dS MLE', scale=alt.Scale(domain=[0, 5], clamp=True)),
    color = "p_value"
).properties(
    width=800,
    height=600)


band = alt.Chart(source).mark_area(opacity=0.7).encode(x='Site',
                                                       y='dN/dS LB', 
                                                       y2='dN/dS UB')

#save(line+band, "FEL_all.png")
line+band
```

```python
source = df_main

line = alt.Chart(source).mark_circle().encode(
    x='Site',
    y = alt.Y('dN/dS MLE', scale=alt.Scale(domain=[0, 5], clamp=True)),
    color = alt.Color(
        'p_value',
        condition=alt.condition(
            alt.datum.p_value <= 0.1,
            alt.value("#E48A3F"),
            alt.value("#BFBEBF")
        ))
        
        
).properties(
    width=800,
    height=600)


band = alt.Chart(source).mark_area(opacity=0.7).encode(x='Site',
                                                       y='dN/dS LB', 
                                                       y2='dN/dS UB')

#save(line+band, "FEL_all.png")
line+band
```
